# IN

```python
df[df['Ozone (ug/m3)'].isin([15.00,12.20,10.60])]
```

```python
query_string = """
SELECT *
FROM df
WHERE "Ozone (ug/m3)" IN (15.00, 12.20, 10.60)
"""
sqldf(query_string, globals())
```

# NOT IN

```python
df[~df['Ozone (ug/m3)'].isin([15.00,12.20,10.60])]
```

```python
query = """
SELECT *
FROM df
WHERE "Ozone (ug/m3)" NOT IN (15.00, 12.20, 10.60);
"""
sqldf(query,globals())
```

# ASCENDING DATA ORDER

```python
df.sort_values(by = ['PM2.5 (ug/m3)','PM10 (ug/m3)'] , ascending=True)
```

```python
query = """
SELECT *
FROM df
ORDER BY "PM2.5 (ug/m3)" ASC, "PM10 (ug/m3)" ASC
"""
sqldf(query,globals())
```

# DESCENDING ORDER

```python
df.sort_values(by = ['PM2.5 (ug/m3)','PM10 (ug/m3)'] , ascending=False)
```

```python
query = """
SELECT *
FROM df
ORDER BY "PM2.5 (ug/m3)" DESC, "PM10 (ug/m3)" DESC
"""
sqldf(query,globals())
```

# AGGREGATION

```python
df.agg( {'CO (mg/m3)': ['count','min','max','mean'] })
```

```python
query = """
SELECT
    COUNT("CO (mg/m3)") AS count,
    MIN("CO (mg/m3)") AS min,
    MAX("CO (mg/m3)") AS max,
    AVG("CO (mg/m3)") AS mean
FROM df;
"""
sqldf(query, globals())
```

# GROUP BY

```python
df.groupby('SO2 (ug/m3)')['RH (%)'].sum()
```

```python
query = """
SELECT
    "SO2 (ug/m3)",
    SUM("RH (%)") AS sum_rh
FROM df
GROUP BY "SO2 (ug/m3)";
"""
sqldf(query, globals())
```

## Group by  with Aggregration

```python
df.groupby('WD (deg)').agg( {'SR (W/mt2)': ['mean','min','max'] })
```

```python
query = """
SELECT
    "WD (deg)",
    AVG("SR (W/mt2)") AS avg_sr,
    MIN("SR (W/mt2)") AS min_sr,
    MAX("SR (W/mt2)") AS max_sr
FROM df
GROUP BY "WD (deg)";
"""
sqldf(query,globals())
```

# JOIN (MERGE)

```python
data1 = {
        'Empid': [1011, 1012, 1013, 1014, 1015],
        'Name': ['John', 'Rahul', 'Rick', 'Morty', 'Tim'],
        'Designation': ['Manager', 'Research Engineer', ' Research Engineer', 'VP', 'Delivery Manager'],
        'Date_of_joining': ['01-Jan-2000', '23-sep-2006', '11-Jan-2012', '21-Jan-1991', '12-Jan-1990']}
Emp_df = pd.DataFrame(data1, columns = ['Empid', 'Name', 'Designation','Date_of_joining'])
Emp_df
```

```python
data2 = {
        'Empid': [1011, 1017, 1013, 1019, 1015],
        'Deptartment': ['Management', 'Research', 'Research', 'Management', 'Delivery'],
        'Total_Experience': [18, 10, 10, 28, 22]}
Dept_df = pd.DataFrame(data2, columns = ['Empid', 'Deptartment', 'Total_Experience'])
Dept_df
```

#INNER JOIN

```python
pd.merge(Emp_df,Dept_df,left_on='Empid',right_on='Empid',how='inner')
```

```python
query = """
SELECT * FROM Emp_df a
 INNER JOIN Dept_df b
 On a.Empid = b.Empid;
 """
sqldf(query,globals())
```

# LEFT JOIN

```python
pd.merge(Emp_df,Dept_df,left_on='Empid',right_on='Empid',how='left')
```

```python
query = """
SELECT * FROM Emp_df a
 LEFT JOIN Dept_df b
 On a.Empid = b.Empid;
 """
sqldf(query,globals())
```

# RIGHT JOIN

```python
pd.merge(Emp_df,Dept_df,left_on='Empid',right_on='Empid',how='right')
```

```python
query = """
select a.Empid,Name,Designation,Date_of_joining,Deptartment,Total_Experience from Dept_df a LEFT JOIN Emp_df b ON  a.Empid = b.Empid;
 """
sqldf(query,globals())
```

#OUTER JOIN

```python
pd.merge(Emp_df,Dept_df,left_on='Empid',right_on='Empid',how='outer')
```

```python
query = """select * from Emp_df a left OUTER JOIN Dept_df b ON  a.Empid = b.Empid;
 """
sqldf(query,globals())
```
